```python
def get_id_from_link(link_name):
    id_value = link_name.split('catalog/', 1)[1]
    id_value = id_value.split('/detail', 1)[0]
    return int(id_value)
```

```python
df_base = df_base.rename(columns={'Наименование': 'name', 'Скидка': 'discount', 'Цена': 'price', 
                                  'Цена со скидкой': 'discount_price', 'Бренд': 'brand', 'Ссылка': 'link'})
```

В столбцах **'Battery', 'MainCameraMP', 'FrontCameraMP'** заменим **NaN** на **0**

```python
df_base[['Battery', 'MainCameraMP', 'FrontCameraMP']] = df_base[['Battery', 'MainCameraMP', 'FrontCameraMP']].replace(np.nan, '0')
```

Преобразуем столбцы **'Battery', 'MainCameraMP', 'FrontCameraMP'** в тип **int**

```python
df_base['memory_size'] = df_base['memory_size'].str.extract('(\d+)').astype(int)
df_base['MainCameraMP'] = df_base['MainCameraMP'].str.extract('(\d+)').astype(int)
df_base['FrontCameraMP'] = df_base['FrontCameraMP'].str.extract('(\d+)').astype(int)
df_base['Battery'] = df_base['Battery'].str.extract('(\d+)').astype(int)
```

Преобразуем столбцы **'weight', 'width', 'height', 'thickness'** в тип **float**

```python
df_base['weight'] = df_base['weight'].str.extract('(\d+)').astype(float)
df_base['width'] = df_base['width'].str.extract('(\d+)').astype(float)
df_base['height'] = df_base['height'].str.extract('(\d+)').astype(float)
df_base['thickness'] = df_base['thickness'].str.extract('(\d+)').astype(float)
df_base['MaxSDMemory'] = df_base['MaxSDMemory'].str.extract('(\d+)').astype(float)
df_base['RAM'] = df_base['RAM'].str.extract('(\d+)').astype(float)
df_base['screen_size_y'] = df_base['screen_size_y'].str.extract('(\d+)').astype(float)
```

```python
df_base[['SIMNumber']] = df_base[['SIMNumber']].replace(np.nan, '1')
```

```python
df_base['SIMNumber'] = df_base['SIMNumber'].str.extract('(\d+)').astype(float)
```
